# Neural network Tuning

```python
# Importing necesary packages: numpy, pandas, GridSearchCV,
#Dense, KerasClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import warnings
warnings.filterwarnings('ignore')
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier 
```

# part 1 - Tuning batch and epoch

```python
# load the dataset for India diabetes
data=pd.read_csv('diabetes.csv')
```

```python
# description de data
data.describe()
```

```python
# dimension de data
data.shape
```

```python
# split into features and target
X=data.iloc[:,0:8]
y=data.iloc[:,8]
```

```python
# Function to create model,for KerasClassifier
def create_my_model():
    
    #defining my model
    mymodel=Sequential()
    mymodel.add(Dense(20,input_dim=8,activation='relu'))
    mymodel.add(Dense(15,activation='sigmoid'))
    mymodel.add(Dense(2,activation='sigmoid'))
    # Compile the model
    mymodel.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return mymodel
```

```python
KerasClassifier?
```

```python
# create model
model=KerasClassifier(build_fn=create_my_model)
# define the grid search parameters
batchSize=[10,20,40,60,80,100]
epochs=[10,30,50]
```

```python
# entrainer le model avec GridSearchCV
paramgrid=dict(batch_size=batchSize,epochs=epochs)
my_grid=GridSearchCV(estimator=model,param_grid=paramgrid)
resultat=my_grid.fit(X,y)
```

```python
resultat.best_score_
resultat.best_params_
```

```python
# summarize results: Best parametres
print(f'best : {resultat.best_params_,resultat.best_score_}')
#print("Best:%f using %S"%(resultat.best_score_,resultat.best_params_))
```

# part 2- Optimizing optimizer

```python
# Function to create model
def create_my_model(optimizer='adam'):
    # create model
    mymodel=Sequential()
    mymodel.add(Dense(20,input_dim=8,activation='relu'))
    mymodel.add(Dense(15,activation='sigmoid'))
    mymodel.add(Dense(2,activation='sigmoid'))
    # Compile the model
    mymodel.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return mymodel
```

```python
# create model
model=KerasClassifier(build_fn=create_my_model,epochs=50,batch_size=10)
```

```python
# define the grid search parameters
optimizer1=['SGD','Adadelta','Adam','Adagrad']
paramgrid=dict(optimizer=optimizer1)
```

```python
# entrainer le model
grid=GridSearchCV(estimator=model,param_grid=paramgrid,n_jobs=-1)
grid_resultat=grid.fit(X,y)
```

```python
# summarize results
grid_resultat.best_params_
```

```python
grid_resultat.best_score_
```
